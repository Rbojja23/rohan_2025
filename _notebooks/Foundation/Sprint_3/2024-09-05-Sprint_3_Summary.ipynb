## Reality Room  
An interactive game and chat area to interact with AI or humans.  
- **Features:**  
  - Post discussions and comments.  
  - Guess if the image is AI-generated or not.  
  - Chatbot for guessing whether the responses are from AI or a human API.  
- [Reality Room Home](https://nighthawkcoders.github.io/flocker_frontend/create_and_compete/realityroom-home)

## Night at the Museum (N@TM) Summary  
We enjoyed presenting our chatroom and game at N@TM. Many attendees were interested in learning more about our project.

## Presentation Overview  
1. Highlighted the main homepage with various posts.  
2. Showcased dynamic chat rooms with unique ID numbers for commenting and chatting.  
3. Presented the interactive game where users guessed if the chatbot responses were from AI or a human.  
4. Demonstrated data and posts saved in a backend SQLite database.
